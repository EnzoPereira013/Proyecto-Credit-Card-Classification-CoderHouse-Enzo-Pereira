## Aplicacion de V de Cramer

```python
import researchpy as rp
import scipy.stats as ss
```

```python
def cramers_v(confusion_matrix):
    """ calculate Cramers V 
    """
    chi2 = ss.chi2_contingency(confusion_matrix)[0]
    n = confusion_matrix.sum()
    phi2 = chi2 / n
    r, k = confusion_matrix.shape
    phi2corr = max(0, phi2 - ((k-1)*(r-1))/(n-1))
    rcorr = r - ((r-1)**2)/(n-1)
    kcorr = k - ((k-1)**2)/(n-1)
    return np.sqrt(phi2corr / min((kcorr-1), (rcorr-1)))
```

```python
df['Target'] = pd.to_numeric(df['Target'], errors='coerce')
df.dropna(subset=['Target'], inplace=True)
df['Target'] = pd.cut(df['Target'], bins=[-np.inf, 0, np.inf], labels=['No', 'Yes'])
```

```python
datos_numericos = df.select_dtypes(include=np.number)
datos_categoricos = df.select_dtypes(include='object')
```

```python
resultados = []
for var1 in datos_categoricos.columns:
    fila = []
    for var2 in datos_categoricos.columns:
        if var1 != var2:
            tabla_contingencia = pd.crosstab(datos_categoricos[var1], datos_categoricos[var2])
            v_cramer = cramers_v(tabla_contingencia.values)
            fila.append(v_cramer) 
        else: 
            fila.append(1.0)
    tabla_contingencia = pd.crosstab(datos_categoricos[var1], df['Target'])
    v_cramer = cramers_v(tabla_contingencia.values)
    fila.append(v_cramer)
    resultados.append(fila)
```

```python
resultados_df = pd.DataFrame(resultados, columns=datos_categoricos.columns.tolist() + ['Target'])
resultados_df.set_index(resultados_df.columns.tolist(), inplace=True)
```

```python
sns.heatmap(resultados, annot=True, xticklabels=datos_categoricos.columns.tolist()+['Target'], yticklabels=datos_categoricos.columns.tolist()+['Target'])
```

#### La V de Cramer es una medida útil para determinar si hay una asociación significativa entre dos variables categóricas y para cuantificar la fuerza de esta asociación. Esta medida varía entre 0 y 1, donde 0 indica que no hay asociación entre las variables y 1 indica una asociación perfecta. Al aplicar esto podemos observar con mayor claridad la asiciacion de las vartiables categoricas de de nuestro DF, cabe sedtacar que las variables con mayor asociacion a la variable objetivo son nivel educativo y la situacion familiar.
