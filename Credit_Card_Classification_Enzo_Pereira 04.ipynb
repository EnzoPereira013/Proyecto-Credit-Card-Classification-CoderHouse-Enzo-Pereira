## Aplicación de V de Cramer

```python
!pip install researchpy
```

```python
import researchpy as rp
import scipy.stats as ss
```

```python
def cramers_v(confusion_matrix):
    """ calculate Cramers V 
    """
    chi2 = ss.chi2_contingency(confusion_matrix)[0]
    n = confusion_matrix.sum()
    phi2 = chi2 / n
    r, k = confusion_matrix.shape
    phi2corr = max(0, phi2 - ((k-1)*(r-1))/(n-1))
    rcorr = r - ((r-1)**2)/(n-1)
    kcorr = k - ((k-1)**2)/(n-1)
    return np.sqrt(phi2corr / min((kcorr-1), (rcorr-1)))
```

```python
df['Target'] = pd.to_numeric(df['Target'], errors='coerce')
df.dropna(subset=['Target'], inplace=True)
df['Target'] = pd.cut(df['Target'], bins=[-np.inf, 0, np.inf], labels=['No', 'Yes'])
```

```python
datos_numericos = df.select_dtypes(include=np.number)
datos_categoricos = df.select_dtypes(include='object')
```

```python
resultados = []
for var1 in datos_categoricos.columns:
    fila = []
    for var2 in datos_categoricos.columns:
        if var1 != var2:
            tabla_contingencia = pd.crosstab(datos_categoricos[var1], datos_categoricos[var2])
            v_cramer = cramers_v(tabla_contingencia.values)
            fila.append(v_cramer) 
        else: 
            fila.append(1.0)
    tabla_contingencia = pd.crosstab(datos_categoricos[var1], df['Target'])
    v_cramer = cramers_v(tabla_contingencia.values)
    fila.append(v_cramer)
    resultados.append(fila)
```

```python
resultados_df = pd.DataFrame(resultados, columns=datos_categoricos.columns.tolist() + ['Target'])
resultados_df.set_index(resultados_df.columns.tolist(), inplace=True)
```

```python
sns.heatmap(resultados, annot=True, xticklabels=datos_categoricos.columns.tolist()+['Target'], yticklabels=datos_categoricos.columns.tolist()+['Target'])
```

#### La V de Cramer es una medida útil para determinar si hay una asociación significativa entre dos variables categóricas y para cuantificar la fuerza de esta asociación. Esta medida varía entre 0 y 1, donde 0 indica que no hay asociación entre las variables y 1 indica una asociación perfecta. Al aplicar esto podemos observar con mayor claridad la asiciación de las variables categoricas de de nuestro DF, cabe destacar que las variables con mayor asociación a la variable objetivo son nivel educativo y la situación familiar.

```python
df_svd = pd.DataFrame(X_svd, columns=['Componente_1', 'Componente_2'])
print(df_svd)
```

## Para este proyecto se ha optado por un modelo de clasificación de random forest por los siguientes motivos: 

## Capacidad para manejar características no lineales:
#### Random Forest es un algoritmo basado en árboles de decisión que tiene la capacidad de capturar relaciones no lineales entre las características y el objetivo. Esto es importante en problemas de clasificación de crédito, donde las relaciones pueden ser complejas y no lineales.

## Robustez frente a datos desbalanceados: 
#### Random Forest es robusto frente a conjuntos de datos desbalanceados, donde la proporción de instancias positivas y negativas en el objetivo puede estar desequilibrada. Esto es relevante en este caso ya que generalmente hay una proporción desigual entre los clientes que cumplen con sus pagos y los que no.

## Reducción del sobreajuste: 
#### Gracias a la técnica llamada "bagging" que combina múltiples árboles de decisión entrenados en diferentes subconjuntos de datos. Esto ayuda a reducir el sobreajuste y mejora la generalización del modelo. En el contexto de clasificación de crédito, donde los datos pueden ser limitados y ruidosos, la reducción del sobreajuste es un aspecto importante a considerar.

## Capacidad de manejar características categóricas: 
#### Random Forest es capaz de manejar características categóricas directamente sin requerir una codificación previa. Esto es beneficioso ya que el conjunto de datos incluye características categóricas como el género, el tipo de empleo o el estado civil, que son relevantes en la clasificación de crédito.

## Interpretabilidad y explicabilidad: 
#### Aunque Random Forest no proporciona una interpretación tan directa como algunos modelos lineales, es posible obtener información sobre la importancia relativa de las características. La importancia de las características se puede obtener mediante la medida de la ganancia de información o el índice Gini en los árboles individuales y luego promediando a través de todos los árboles del bosque. Esta información puede ser útil para entender que características son más relevantes para la clasificación de crédito.

```python
X_train, X_test, y_train, y_test = train_test_split(df_svd, y, test_size=0.3, random_state=42)
```

```python
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)
y_pred = rf.predict(X_test)
print("Exactitud (Accuracy):", accuracy_score(y_test, y_pred))
print("\nReporte de clasificación:")
print(classification_report(y_test, y_pred))
print("\nMatriz de confusión:")
print(confusion_matrix(y_test, y_pred))
```

#### Luego de probar diferentes combinaciones de hiperparámetros no se logro un aumento significativo de la precisión asi que se decidio realizar una búsqueda de los mejores hiperparámetros :

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
param_grid = {
    'n_estimators': [100, 200, 300, 400, 500, 600, 700, 800, 1000],
    'max_depth': [5, 10, 15, 20, 25, 30, 35, 40, 45],
    'min_samples_split': [2, 5, 10, 12, 15, 17, 20],
    'min_samples_leaf': [1, 2, 4, 6, 8, 10, 12],
    'max_features': ['sqrt', 'log2']
}
rf = RandomForestClassifier(random_state=42)
random_search = RandomizedSearchCV(estimator=rf, param_distributions=param_grid, n_iter=10, cv=5, scoring='accuracy', random_state=42)
random_search.fit(X_train, y_train)
best_params = random_search.best_params_
print("Mejores hiperparámetros:", best_params)
best_rf = RandomForestClassifier(**best_params, random_state=42)
best_rf.fit(X_train, y_train)
y_pred = best_rf.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print("Exactitud (Accuracy) con mejores hiperparámetros:", accuracy)
```

#### Luego de buscar los mejores hiperparámetros se los aplico al modelo logrando una precisión del 87%
