# Visualizaciones: 

##Relacion entre desempleo y riesgo

```python
tabla_contingencia_Desempleo_incumplimiento = pd.crosstab(df["Unemployed"], df["Target"])
print(tabla_contingencia_Desempleo_incumplimiento)
```

```python
sns.heatmap(tabla_contingencia_Desempleo_incumplimiento, annot=True, cmap="YlGnBu")
plt.show()
```

#### En el Hetamap se puede observar que el estar desemepleado no representa un riesgo de incumplimiento, ya que la mayor agruoacion esta en no desempleado y bajo riesgo.

```python
from scipy.stats import pearsonr
```

```python
corr, p_val = pearsonr(df['Unemployed'].astype(int), df['Target'].astype(int))
print("La correlación es:", corr)
print("El valor p es:", p_val)
```

#### Interpretacion: al estar el valor de la correlacion cercano a 0 indica que no existe correlacion entree estas variables, aunque el resultado del valor P indica que hay evidencia suficiente para rechazar la hipótesis nula de que no hay correlación entre las variables.
