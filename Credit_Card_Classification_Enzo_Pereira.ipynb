## Dataset: Credit Card Classification.

#### La puntuación crediticia es un método común de control de riesgos en la industria financiera. Utiliza la información personal y los datos presentados por los solicitantes de tarjetas de crédito para predecir la probabilidad de futuros incumplimientos y préstamos de tarjetas de crédito. El banco puede decidir si emite una tarjeta de crédito al solicitante. Los puntajes de crédito pueden cuantificar objetivamente la magnitud del riesgo. 
#### En términos generales, la puntuación de crédito se basan en datos históricos. Una vez que se encuentra con grandes fluctuaciones económicas. Los modelos anteriores pueden perder su poder predictivo original.



## Contexto Comercial: 

#### Para las empresas que prestan servicios finiancieros el incumlimiento de pagos por parte de los usuarios de lineas de credito podria representan perdidas importantes, teniendo esto en cuenta es impiortante la creacion de un modelo que puda prevenir estos posibles incumplimientos con el fin de disminuir estos riesgos. 

## Contexto Analitico:

#### Se cuenta con un Dataset de diferentes solicitantes  de tarjetas de credito que contiene diversas variables con las cuales se pretende entrenar al modelo con el fin de que pueda detectar los riesgos existentes.

#Visualizaciones de datos, segunda entrega de proyecto. 

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

##Graficas en matplotlib:

```python
mpl.style.use ('bmh')
```

#Grafica de distribucion de la muestra segun genero:

```python
gender_counts = df['Gender'].value_counts()
```

```python
fig, ax = plt.subplots()
ax.bar(gender_counts.index, gender_counts.values)
ax.set_title('Distribución de la poblacion segun genero')
ax.set_xlabel('Género')
ax.set_ylabel('Numero de individuos')
ax.set_xticks([0, 1])
ax.set_xticklabels(['Masculino', 'Femenino'])
ax.legend(['Cantidad'])
plt.show()
```

#### En esta grafica podemos observar la distribucion de la muestra segun el genero, como se puede apreciar hay mayyor cantidad de individuos del genero masculino, practicamente duplicando el numero del genero femenino.

#Tipos de ingreso de la muestra:

```python
income_sources = df['Income_type'].value_counts()
```

```python
fig, ax = plt.subplots()
ax.pie(income_sources.values, labels=income_sources.index, autopct='%1.1f%%')
ax.set_title('Fuentes de ingresos')
```

#### En esta grafica podemos observar las diferentes fuentes de ingresos de los individuos de la muestra, cabe resaltar que la mayoria de los solicitantes son trabajadores.

#Distribucion de la poblacion segun presencia o no de riesgo de incumplimiento

```python
target_counts = df['Target'].value_counts()
```

```python
fig, ax = plt.subplots()
ax.bar(target_counts.index, target_counts.values)
ax.set_title('Distribución de la poblacion segun riesgo')
ax.set_xlabel('Riesgo')
ax.set_ylabel('Numero de individuos')
ax.set_xticks([0, 1])
ax.set_xticklabels(['Sin riesgo', 'Presenta riesgo'])
ax.legend(['Cantidad'])
plt.show()
```

#### En esta grafica se compara la cantidad de individuos que presentan riesgo de incumplimiento de pago de su targeta de credito en comparacion a los que no presentan riesgo, podemos observar que en su mayoria no presentan riesgo. 

##Graficas en seaborn:

```python
edu_levels = df['Education_type'].value_counts()
```

```python
sns.set_style('whitegrid')
sns.countplot(x='Education_type', data=df, order=edu_levels.index)
plt.title('Niveles educativos de la muestra')
plt.xlabel('Nivel educativo')
plt.ylabel('Numero de individuos')
plt.xticks(rotation=90)
```

#### En la presente grafica se comparan los diferentes niveles educativos de la muestra, se puede observar una prevalencia importante del nivel secundario. 

```python
civil_status = df['Family_status'].value_counts()
sns.set_style('whitegrid')
fig, ax = plt.subplots()
ax.pie(civil_status.values, labels=civil_status.index, autopct='%1.1f%%')
plt.title('Estados civiles de la muestra')
plt.show()
```

#### En esta grafica podemos observar los diferentes estados civiles de la poblacion, se puede destacar que la mayoria se encuentra casado, seguido por soltero y union civil. 

```python
edu_salary = df[['Education_type', 'Total_income']]
sns.set_style('whitegrid')
sns.relplot(x='Education_type', y='Total_income', data=edu_salary)
plt.title('Relación entre nivel educativo y salario')
plt.xlabel('Nivel educativo')
plt.ylabel('Salario')
plt.xticks(rotation=90)
plt.show()
```

#### En esta grafica podemos observar la relacion entre entre los ingresos totales y el nivel educativo de los individuos, se puede ver que la educacion superior es quien poresenta un mayor ingreso total en relacion a los demas niveles.
